#### Details for the columns present in the dataset 

**'YEAR':** Represents the year in which the data was recorded or collected.

**'Borough':** Indicates the borough or administrative region where the HIV/AIDS diagnoses were reported. In many cities, such as New York City, a borough is a separate administrative division within the city.

**'Neighborhood (U.H.F)':** Refers to the specific neighborhood or area within a city, often categorized based on the UHF (United Hospital Fund) designation. UHF neighborhoods are a way of organizing and categorizing different regions within a city for administrative or analytical purposes.

**'SEX':** Represents the gender or sex of the individuals diagnosed with HIV/AIDS. This column might include categories such as male, female, or other gender identities depending on the available data.

**'RACE/ETHNICITY':** Indicates the racial or ethnic background of the individuals diagnosed with HIV/AIDS. This column might contain categories representing different racial groups or ethnicities, such as White, Black/African American, Hispanic/Latino, Asian, etc.

**'TOTAL NUMBER OF HIV DIAGNOSES':** Refers to the overall count or total number of individuals diagnosed with HIV within a specific area, borough, or neighborhood during a particular time period (usually a year).

**'HIV DIAGNOSES PER 100,000 POPULATION':** Represents the rate of HIV diagnoses per 100,000 individuals in the population. It's a standardized measure to compare HIV diagnosis rates across different areas or populations with varying sizes.

**'TOTAL NUMBER OF CONCURRENT HIV/AIDS DIAGNOSES':** Indicates the total count of individuals diagnosed with both HIV and AIDS concurrently within a specific area or population during a certain period.

**'PROPORTION OF CONCURRENT HIV/AIDS DIAGNOSES AMONG ALL HIV DIAGNOSES':** Represents the percentage or proportion of individuals diagnosed with both HIV and AIDS concurrently among all diagnosed cases of HIV within a particular region or population.

**'TOTAL NUMBER OF AIDS DIAGNOSES':** Refers to the overall count or total number of individuals diagnosed with AIDS within a specific area, borough, or neighborhood during a particular time period (usually a year).

**'AIDS DIAGNOSES PER 100,000 POPULATION':** Similar to the HIV diagnosis rate, this column represents the rate of AIDS diagnoses per 100,000 individuals in the population. It's a standardized measure to compare AIDS diagnosis rates across different areas or populations with varying sizes.
